```python
#Standard scaler

# backup
df_ready = df.copy() # since undo is not there in python

# importing other important libraries
from sklearn.preprocessing import StandardScaler

# Initialization
Scaler = StandardScaler()

# Making a list of numerical columns
num_cols=['CREDIT_SCORE' ,'CUSTOMER_RATING','Monthly_Income','DTI',
          'Total_Experience','months_current_job','Existing_Customer_Since_When'
          ,'Avg Balance','Count_minimum_balance_Total','Savings_Account_Total',
          'Savings_Account_Cable Bill_Total','Savings_Account_Dining_Total','Savings_Account_Electricity Bill_Total','Savings_Account_Entertainment_Total',
          'Savings_Account_Gas Bill_Total','Savings_Account_Gasoline/Fuel_Total','Savings_Account_Grocery_Total','Savings_Account_Internet Bill_Total',
         'Savings_Account_Merchant_Total','Savings_Account_Mobile Recharge_Total','Savings_Account_Pharmacy/Healthcare_Total','Savings_Account_Shopping_Total',
          'Savings_Account_Telephone Bill_Total','Savings_Account_Ticket Booking_Total','Savings_Account_Travel_Total','Savings_Account_Water Bill_Total',
         'Savings_Account_Wellness and Fitness_Total','Savings_Churn_Predicted_Label']

# scaling
df_ready[num_cols] = Scaler.fit_transform(df_ready[num_cols])
df_ready.head()
```

```python
# dealing with categorical columns

# list of categorical columns
cat_cols = ['Products','Organisation_type','Marital_Status','Type_of_employment','Education','Balance_Trend_Segment']

# Encoding
df_ready = pd.get_dummies(data = df_ready, columns = cat_cols)

# encoding other yes no values of dataset
df_ready['IsOnlineBankingUser'] = df_ready['IsOnlineBankingUser'].apply(lambda x : 1 if x =='yes' else 0)
df_ready['IsSmsBankingUser'] = df_ready['IsSmsBankingUser'].apply(lambda x : 1 if x =='yes' else 0)
df_ready['IsChequeBookUser'] = df_ready['IsChequeBookUser'].apply(lambda x : 1 if x =='yes' else 0)
df_ready['Is_app_user'] = df_ready['Is_app_user'].apply(lambda x : 1 if x =='yes' else 0)
df_ready['Is_Life_Insurance_Holder'] = df_ready['Is_Life_Insurance_Holder'].apply(lambda x : 1 if x =='yes' else 0)

df_ready.head()
```

```python
# printing the shape of data
print(df_ready.shape)
```

```python
df['Life-Insurance_Cross-Sell prediction'].value_counts()
```

```python
# seperating input and output features

# input feature
features = df_ready.drop('Life-Insurance_Cross-Sell prediction',axis = 1)

# target
target = df_ready['Life-Insurance_Cross-Sell prediction']
```

```python
# Train Test Split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(features, target, 
                                                    test_size = 0.2, random_state = 7, stratify=target)
```

```python
# support vector machine
```

```python
#model
from sklearn.svm import SVC
svm = SVC(kernel = 'rbf')

# fitting the model
svm.fit(X_train, y_train)
```

```python
# Predocting from the model
y_pred = svm.predict(X_test)
```

```python
# Evaluatio
from sklearn.metrics import accuracy_score
print('Accuracy:', accuracy_score(y_test, y_pred))
```

```python
print(classification_report(y_test, y_pred))
```

```python
confusion_matrix(y_test,y_pred)
```
